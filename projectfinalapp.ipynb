```python
%%writefile app.py
import streamlit as st
from tensorflow.keras.models import load_model
import numpy as np
from PIL import Image

# Load the trained model (replace with your actual model loading code)
model = load_model('/content/drive/MyDrive/Project final/image_classificationdiseasesfinal.h5')

# Define class labels (replace with your actual class labels)
class_labels = [
    'Potato___Early_blight',
    'Potato___Late_blight',
    'Potato___healthy',
    'Strawberry___Leaf_scorch',
    'Strawberry___healthy',
    'Tomato___Early_blight',
    'Tomato___Late_blight',
    'Tomato___healthy'
]

# Streamlit app
st.title('Plant Disease Detection App')

# Upload image through Streamlit
uploaded_file = st.file_uploader("Choose a plant image...", type=["jpg", "jpeg", "png"])

if uploaded_file is not None:
    # Preprocess the image
    image = Image.open(uploaded_file)
    st.image(image, caption='Uploaded Image.', use_column_width=True)

    # Resize and preprocess the image for the model
    image = image.resize((180, 180))
    image_array = np.asarray(image)
    # image_array = image_array / 255.0  # Normalize the pixel values
    image_array = np.expand_dims(image_array, axis=0)  # Add batch dimension

    # Make predictions
    predictions = model.predict(image_array)

    # Get the predicted class index with the highest probability
    predicted_class_index = np.argmax(predictions[0])

    # Display the result using if-else if statements
    if predicted_class_index == 0:
        st.write("Predicted Class: Potato___Early_blight")
        st.write("Syndrome:")
        st.write("Dark brown to black lesions on lower leaves.")
        st.write("Small, dark spots with concentric rings on leaves.")
        st.write("Yellowing between lesions, leading to overall discoloration.")
        st.write("Withering, drying, and premature death of affected leaves.")
        st.write("Treatment:")
        st.write("Organic treatment:")
        st.write("Use products based on Bacillus subtilis or Copper-based fungicides")
        st.write("Chemical treatment:")
        st.write("Use the product that contains chemicals like Azoxystrobin, Pyraclostrobin, Mancozeb, ziram, Fenamidone")

    elif predicted_class_index == 1:
        st.write("Predicted Class: Potato___Late_blight")
        st.write("Syndrome:")
        st.write("Small, dark, water-soaked lesions on leaves.")
        st.write("White, fuzzy growth (sporulation) on the undersides of infected leaves.")
        st.write("Curling and wilting, leading to a burnt appearance.")
        st.write("Darkening of veins on infected leaves.")
        st.write("Treatment:")
        st.write("Apply a copper-based fungicide (2 oz/ gallon of water) every 7 days or less")
        st.write("Used as a foliar spray")
        st.write("Mix 2 tsp/ gallon of water and spray at transplant")
        st.write("when direct-seeded crops are at 2-4 true leaf")
    elif predicted_class_index == 2:
        st.write("Predicted Class: Potato___healthy")
        st.write("Your Plant Is  Healthy")
    elif predicted_class_index == 3:
        st.write("Predicted Class: Strawberry___Leaf_scorch")

    elif predicted_class_index == 4:
        st.write("Predicted Class: Strawberry___healthy")
        st.write("Your Plant Is  Healthy")

    elif predicted_class_index == 5:
        st.write("Predicted Class: Tomato___Early_blight")
        st.write("Syndrome:")
        st.write("Dark brown to black lesions on lower leaves.")
        st.write("Small, dark spots with concentric rings on leaves.")
        st.write("Yellowing between lesions, leading to overall discoloration.")
        st.write("Withering, drying, and premature death of affected leaves.")
        st.write("Treatment:")
        st.write("Organic treatment:")
        st.write("Use products based on Bacillus subtilis or Copper-based fungicides")
        st.write("Chemical treatment:")
        st.write("Use the product that contains chemicals like Azoxystrobin, Pyraclostrobin, Mancozeb, ziram, Fenamidone")

    elif predicted_class_index == 6:
        st.write("Predicted Class: Tomato___Late_blight",color='r')
        st.write("Syndrome:")
        st.write("Small, dark, water-soaked lesions on leaves.")
        st.write("White, fuzzy growth (sporulation) on the undersides of infected leaves.")
        st.write("Curling and wilting, leading to a burnt appearance.")
        st.write("Darkening of veins on infected leaves.")
        st.write("Treatment:")
        st.write("Apply a copper-based fungicide (2 oz/ gallon of water) every 7 days or less")
        st.write("Used as a foliar spray")
        st.write("Mix 2 tsp/ gallon of water and spray at transplant")
        st.write("when direct-seeded crops are at 2-4 true leaf")

    elif predicted_class_index == 7:
        st.write("Predicted Class: Tomato___healthy")
        st.write("Your Plant Is  Healthy")
    # Calculate accuracy by taking the maximum probability
    accuracy = np.max(predictions[0]) * 100
    st.write(f"Accuracy: {accuracy:.4f}")
```
